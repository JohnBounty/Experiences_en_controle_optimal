On considère le rendez-vous de deux vaisseaux spatiaux au voisinage de la Terre (supposée fixe) sous les hypothèses suivantes :
* Le vaisseau 1 est désigné comme passif et on lui confère une trajectoire circulaire (centrée sur la Terre) à vitesse angulaire constante $\omega$ connue (Képlerienne).
* Le vaisseau 2, de position initiale connue, est considéré comme actif : doté d’un système générant une poussée $v(t) = (v_1(t), v_2(t))$ , on cherche à lui faire rencontrer le vaisseau 1 au temps fixé T.
* On traite ici le problème en dimension deux, dans le plan $(O, X, Y )$ centré sur la terre et on suppose que les vaisseaux sont des corps ponctuels.

Sous ces hypothèses, la dynamique du vaisseau 2 dans le repère $\mathcal{R}=\left(M, \vec{e}_{x}, \vec{e}_{y}\right)$ vérifie les équations de Hill Clohessy-Wiltshire : en notant $x(t), x'(t), y(t), y'(t)$ les positions et vitesses (relatives) du vaisseau 2 dans le repère $\mathcal{R}$ et $X(t) = (x(t),x'(t),y(t),y'(t))^T$, on a

$$
\begin{cases}{\dot{X}(t)=A X(t)+B v(t)} \\ {X(0)=X_{0}}\end{cases}, \quad
\text{avec  }
A=\begin{pmatrix}{0} & {1} & {0} & {0} \\ {3 \omega^{2}} & {0} & {0} & {2 \omega} \\ {0} & {0} & {0} & {1} \\ {0} & {-2 \omega} & {0} & {0}\end{pmatrix} \quad
\text{et } B=\begin{pmatrix}{0} & {0} \\ {1} & {0} \\ {0} & {0} \\ {0} & {1}\end{pmatrix}
$$

où $\omega$ est la vitesse angulaire définie par $\omega = 2\pi/\tau$ où $\tau = 5480$ s désigne la période de révolution
pour la station spatiale internationale.

On souhaite que les deux vaisseaus se rencontrent au temps $T$ avec une vitesse relative nulle lors de l’impact. On modélise ce problème sous la forme
$$
\mathcal{(P)} \quad\quad \inf _{v \in L^{2}(] 0, T[)} J(v), \quad \text{où  } J(v)=\frac{\varepsilon}{2} \int_{0}^{T} v(t)^{2} d t+\frac{1-\varepsilon}{2}\|X(T)\|^{2}, \\ 
\varepsilon \in] 0,1[ \quad \text{est un paramètre fixé.}
$$

On cherchera donc ici, à trouver la poussée optimale qui minimise la fonctionnelle $J$.

## I - La pertinence physique du problème

Pour commencer, discutons de la signification de la fonction de coût $J(v)$. Le premier terme $\int_{0}^{T} v(t)^{2}$ représente la puissance totale, par conséquent, le combustible, qui est dépensé de $0$ à $T$. Le second terme ne nous pénalisera que dans le cas où $X(T)$ n'est pas $0$, ce qui signifie que la position, la vitesse ou les deux au moment $T$ n'étaient pas nulles. Cela se traduit directement par le fait que soit le vaisseau n'a pas atteint sa destination, soit il y est arrivé à grande vitesse, ce qui constitue essentiellement une collision.

Quant au système d'équations différentielles proposé, il est décrit par le changement de position $(x(t), y(t))$ et de vitesse de déplacement du vaisseau $(x'(t), y'(t))$ plus le contrôle $v(t) = (v_1(t), v_2(t))$ que nous sommes en mesure d'effectuer. En tenant compte de la matrice $B$, on affirme qu'il est seulement dans notre capacité d'influencer directement la vitesse du vaisseau, qui semble correspondre au processus du mouvement tel qu'il est dans le monde réel.

Pour conclure, en tenant compte de ce qui précède, on confirme la pertinence physique du modèle proposé.

## II - Calcul de gradient de $J$

En écrivant l'expression pour le gradient de $J(v)$ dans la direction $h$, on obtient:
$$
    \nabla J(v)\cdot h = \varepsilon \int_{0}^T v(t)h(t)dt + (1-\varepsilon) X(T) \hat{X}(T) \\
    \text{où } \hat{X} \text{ résout } \\
    (1)\quad\quad \begin{cases}\hat{X}'(t) = A\hat{X}(t) + Bh(t) \\ 
    \hat{X}(0) = 0\end{cases}
$$
On veut mettre le deuxième terme $ X(T) \hat{X}(T)$ sous la propre forme: $\int_0^T h(t)\cdot truc$. \\
Pour atteindre cet objectif, on introduit un état adjoint $P$ $-$ solution de 
$$
    (2)\quad\quad \begin{cases}P'(t) = -A^TP(t)  \\ 
    P(T) = \quad?\end{cases}
$$


On commence par multiplier $(1)$ par $P$ et par l'intégrer:
$$
\int_0^T \hat{X}'(t)P(t)dt  = \int_0^T A\hat{X}(t)P(t)dt + \int_0^T Bh(t)P(t)dt \\
\text{Par IPP: } 
\int_0^T \hat{X}'(t)P(t)dt  = -\int_0^T \hat{X}(t)P'(t)dt + [\hat{X}(t)P(t)]_0^T \\
$$
Alors, en également considerant que $P'(t) = -A^TP(t)$ en $(2)$, on obtient:

$$
\int_0^T \hat{X}(t)A^TP(t)dt + \hat{X}(T)P(T) = \int_0^T \hat{X}(t)A^TP(t)dt + \int_0^T Bh(t)P(t)dt \\ 
$$
ce, qui nous permet d'écrire l'identité de dualité:

$$
\underline{\hat{X}(T)P(T) = \int_0^T h(t)\cdot B^TP(t)dt}
$$
On choisi $P(T) = X(T)$ et, donc, $truc = B^TP(t)$, et alors:

$$
X(T)\hat{X}(T) = \int_0^T h(t)\cdot B^TP(t)dt
$$

Finalement, le gradient de $J$ s'écrit comme:
$$
\boxed{\quad\nabla J(v) = \varepsilon v(t) + (1-\varepsilon)B^TP(t)\quad} \\
$$
où $P$ est l'état adjoint de (1), solution de:
$$
    (2)\quad\quad \begin{cases}P'(t) = -A^TP(t)  \\ 
    P(T) = X(T)\end{cases}
$$

## III - L'existence et l'unicité de la solution

*On veut montrer que le problème ci-dessus possède une unique solution.

1) On va d'abord montrer que $J(v)$ est $\alpha$-convexe.

$$
    J(v) \text{    est }\alpha\text{-convexe} \iff \langle \nabla J(u) - \nabla J(v), u-v\rangle \\
$$

On calacul la différence $\nabla J(u) - \nabla J(v) = \varepsilon(u - v)$, et on obtient:

$$
   \langle \nabla J(u) - \nabla J(v), u-v\rangle = \langle \varepsilon(u - v), u-v\rangle = \varepsilon ||u-v||^2
$$

Dans notre cas $\varepsilon  > 0$, alors on constate que $J(v)$ est $\alpha$-convexe.

2) On veut aussi que $J(v)$ soit continue.
$$
|J(v) - J(v_0)| = |\frac{\varepsilon}{2} \int_0^T (v^2 - v_0^2) dt| = 
\frac{\varepsilon}{2} \left(||v||^2_{L_2} - ||v_0||^2_{L_2} \right) = 
\frac{\varepsilon}{2} \left(||v||_{L_2} - ||v_0||_{L_2} \right)\left(||v||_{L_2} + ||v_0||_{L_2} \right) \\
\text{ On note}\quad C = \frac{\varepsilon}{2} \left(||v||_{L_2} + ||v_0||_{L_2} \right), \quad \text{et par l'inégalité triangulaire inverse, on obtient : } \\
|J(v) - J(v_0)|  \leq C\left(||v - v_0||_{L_2} \right) \\
$$

Alors, $J(v)$ est continue.

Maintenant, on peut appliquer le théorème d'existence comme $v \in U = L_2\{0, T\}$ qui est convexe, fermée et non vide si $T>0$ et constater que le pb $\mathcal{(P)}$ possède une unique solution.



Énonçons également la CNS d'optimalité. Si le pb $\mathcal{(P)}$ possède une unique solution $u^* \in U$, alors, $u^*$ vérifie l'inéquation d'Euler:

$$
    \langle \nabla J(u^*), v - u^*\rangle \geq 0, \quad\quad \forall v \in U
$$

En prenant $v = u^* + h$, on obtient:
$$
    \langle \nabla J(u^*), h\rangle \geq 0, \quad\quad \forall h \in U
$$

En prenant $v = u^* - h$, on obtient:

$$
    -\langle \nabla J(u^*), h\rangle \geq 0, \quad\quad \forall h \in U
$$

Ainsi, on conclut, que:

$$
    -\langle \nabla J(u^*), h\rangle = 0, \quad\quad \forall h \in U, \\
    \nabla J(u^*) \in U^\perp
$$



## IV - L'algorithme de la résolution numérique

Pour résoudre ce problème numériquement on va utiliser la méthode de gradient à pas constant. On introduit deux critères d'arrêt:
1. Le nombre d'itérations $i$ a atteint le nombre maximum d'itérations $\text{"Iterations"}$
2. La solution de l'itération $i+1$ n'est pas suffisamment différente de celle de l'itération $i$: $ ||u_{i+1} - u_i|| \leq \varepsilon_{tol}$, où $\varepsilon_{tol}$ est le niveau de tolérance donné.

De plus, avant de démarrer la boucle, il faut initialiser les variables suivantes :

- $X_0$ - l'état initial
- $dt$ - le pas de discrétisation
- $N$ - le nombre des pas de discrétisation
- $\rho$ - la vitesse de descente du gradient

Alors, l'itération de l'algorithme de gradient à pas constant est tel que montré ci-dessous:

>>> $\text{ Iteration } i:$
>>> $$X = \text{solve_Heun}(u_i, X_0, dt, N)$$
>>> $$P = \text{solve_Heun}(0, X(T), dt, N)$$
>>> $$dJ = -\nabla J(u_i, P)$$
>>> $$u_{i+1} = u_i + \rho dJ$$
>>> $$ i = i + 1$$

Il faut remarquer, qu'à chaque itération on résout deux équations différentielles: sur l'état $X$ et sur l'adjoint $P$. Cela se fait en appelant la fonction $\text{solve_Heun}$, qui résout ces équations numériquement en utilisant la méthode de Heun.

### Résolution d'un système

```python
# les matrices A & B
tau = 5480
little_omega = 2*np.pi/tau
print(little_omega)
A = np.array([[0, 1, 0, 0],[3*little_omega**2, 0, 0, 2*little_omega], 
                [0, 0, 0, 1],[0, -2*little_omega, 0, 0]])
B = np.array([[0, 0], [1, 0],
                [0, 0], [0, 1]])

# discretisation
T = 1
N = 100
dt = T/N
```

```python
def solve_Heun(v, y0, dt, N, mode):
    def rhs_et(y, v):
        return A@y + B@v
    
    def rhs_adj(y, v):
        # on résout pour le temps inversé
        # p = p(s) = p(T-t)
        # alors "-" disparaît
        return A.T@y

    y=np.zeros((4, N))

    # état ou adjoint
    if mode == 1:
        rhs = rhs_et
    if mode == 2:
        rhs = rhs_adj

    y[:, 0] = y0
    # résolution
    for i in range(1, N):
        y_hat = y[:, i-1] + dt * rhs(y[:, i-1], v[:, i-1]) 
        y[:, i] = y[:, i-1] + (dt / 2) * (rhs(y[:, i-1], v[:, i-1]) + rhs(y_hat, v[:, i]))

    return y
```

```python
# testing
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
v=np.ones((2, N))
y0 = 0
y = solve_Heun(v, y0, dt, N, mode=1)
ax1.set_title("Etat")
ax1.plot(y[0])
ax1.plot(y[2]);

y=np.zeros((4, N))
v=np.ones((2, N))
y0 = 0
y = solve_Heun(v, y0, dt, N, mode=2)
ax2.set_title("Adjoint")
ax2.plot(y[0])
ax2.plot(y[2]);
```

### Résolution de $\inf _{v} J(v)$

```python
eps = 0.1
def J(v, XT, N):
    return (eps/2)*(1/N)*np.sum(v**2) + ((1-eps)/2)*np.linalg.norm(XT)**2

def dJ(v, p):
    return eps*v + (1-eps)*np.transpose(B)@p
```

```python
# test
vp = np.ones((2,100))
pt = np.ones((4, 100))*5
dJ(vp, pt).shape
```

```python
def inf_J(lr, u0, N, dt, X0, J, dJ, epsilon=1e-5, print_epoch=25, 
          Iterations = 4004, plot = False,
           cooling_epoch=1000000, heating_epoch = 1000000):

    # initialisation
    # contrôle
    u_old = np.stack([np.random.normal(size=N), np.random.normal(size=N)])
    u_new = u0
    # état
    x = np.zeros((4, N))
    # adjoint
    p = np.zeros((4, N))
    i = 0

    routeF = np.array([J(u0, x[:,-1], N)])
    routeX = np.array([u0])
    
    try:
        # la boucle principal
        while np.linalg.norm(u_old-u_new) > epsilon and i < Iterations:
            i+=1
            u_old = np.copy(u_new)
            # resolution du (1)
            x = solve_Heun(u_old, X0, dt, N, mode=1)
            # resolution du (2)
            p = solve_Heun(np.zeros_like(u_old), x[:, -1], dt, N, mode=2)
            p = p[:, ::-1]
            
            # pas du gradient
            d = -1*dJ(u_old, p)
            u_new = u_old + lr*d
            
            routeF = np.append(routeF, J(u_new, x[:, -1], N))
            routeX = np.concatenate((routeX, [u_new]))      

            # les expériences avec
            # changement de la vitesse de descente du gradient
            if i%cooling_epoch == 0:
                lr /= 10

            if i%heating_epoch == 0:
                lr *= 10
            
            if i%print_epoch == 0:
                print("[Minimization epoch {:d}] >>> target function value is {:.4f}.".format(i, routeF[i]))
            

    except KeyboardInterrupt:
        print("The algorithm was stopped at epoch {:d}.".format(i))
    
    # min_iter = np.argmin(routeF[1:])
    # u_opt = routeX[min_iter]
    u_opt = u_new.copy()
    # plot pour voir la route du gradient
    if plot:
        i = len(routeF)
        # on calcule les erreurs sur chaque pas
        errors = np.zeros(i)
        for k in range(i-1):
            errors[k] = np.linalg.norm(routeX[k] - u_opt)      
        
        
        xx = np.arange(0, i, 1)
        # print("F: ", routeF.shape)
        # print("xx: ", xx.shape)
        # print("Errs: ", errors.shape)
         
        fig,axs = plt.subplots(2, 1, figsize = (12, 6), dpi =128, sharex = True)
        axs[0].set_title("Valeurs de la fonctionnelle"
                         " évaluée à chaque pas de gradient à pas fixe")
        axs[0].plot(xx, routeF, "--", color = "coral", linewidth = 2)
        #axs[0].plot(min_iter, routeF[min_iter], "*", markersize = 5, color = "orange")
        axs[1].set_title("Erreurs - Log scale")
        axs[1].set_yscale("log")
        axs[1].plot(xx, errors, "--", color = "teal", linewidth = 2)
        plt.xlabel("Iteration");
    
    return u_opt, routeF, routeX, x, p
```

```python
def point_translation(X, t):
    
    station = np.array([np.cos(t), np.sin(t)])
    tang = np.array([-np.sin(t), np.cos(t)])
    normal = np.array([np.cos(t), np.sin(t)])
    P = np.stack([tang, normal])
    theta = np.arctan2(tang[1,:], tang[0,:])
    RM = np.array([[np.cos(theta), -np.sin(theta)],[np.sin(theta), np.cos(theta)]])

    translated_X = station + X
    new_X = np.zeros_like(X)

    for i in range(X.shape[1]):
        new_X[:, i] = RM[:, :, i]@translated_X[:, i]

    return new_X
```

##VI - Une amélioration de l’algorithme de gradient

Une amélioration possible pour l'algorithme serait d'ajouter une technique de refroidissement/chauffage : diminuer ou augmenter la valeur de $\rho$ après une itération donnée ou en fonction de $i$, en fonction de $|| u_{i+1} - u_i ||$ ou  $|| J(u_{i+1}) - J(u_i) ||$.

Une version de cette approche a été mise en œuvre et testée. Il a été décidé de multiplier rho par une constante à chaque échauffement_époque, si elle est donnée, ou de diviser par une constante à chaque refroidissement_époque, si elle est donnée. 
Par exemple, si on voit qu'à un moment donné $J$ cesse de s'améliorer, on peut essayer de "chauffer" un peu le $\rho$ pour avoir plus d'impact sur la petite valeur du gradient.  Cependant, s'il est vrai que le minimum local a été dépassé, on peut essayer de "refroidir" le $\rho$ à un moment donné.

On peut également faire remarquer qu'au lieu d'exercer une influence sur rho, on peut essayer de s'échapper du plateau en ajoutant une petite perturbation à la valeur actuelle de $u$, mais cela peut tout aussi bien être risqué. Une telle approche n'a pas été testée.

```python
def integrate(f, a, b, nb_points=1000, mode='approx'):
    # la fonction calcule integrale(f) de _a à _b
    # f est une fonction, qui prend une argument t
    assert a<b, "a <= b"
    t = np.linspace(a, b, nb_points)

    if mode == 'approx':
        value = np.sum(f(t))*(b-a)/nb_points
    if mode == 'trap':
        value = (b-a)*(f((a+b)/2))
    return value

def grad_1d(f, v, perturbation = 1e-4):
    # calcul numérique du gradient d'une fonction d'une seule variable(R^1)
    return (f(v + perturbation) - f(v - perturbation))/(2*perturbation)

def gradient(func, V, epsilon = 0.000001):
    N = len(V)
    grad = np.zeros(N)
    Veps = np.zeros(N)
    for i in range(N):
        Veps[i] = epsilon
        grad[i] = (func(V + Veps) - func(V - Veps))/(2*epsilon)
        Veps[i] = 0
    return grad
```

```python
a = 0 
T = 10
X0 = np.array([42, 0, 20, 0])
# par formule de Duhamel
def X_t(t, V, little_omega=0.44):
    # les matrices A & B
    A = np.array([[0, 1, 0, 0],[3*little_omega**2, 0, 0, 2*little_omega], 
                  [0, 0, 0, 1],[0, -2*little_omega, 0, 0]])
    B = np.array([[0, 0], [1, 0],
                 [0, 0], [0, 1]])

    def inside_integral(s):
        return expm((t-s)*A)@B@V

    return expm(A*t)*X0 + integrate(inside_integral, 0, t, mode="trap")

def J(v, epsilon=0.777):
    def inside_integral(v):
        return v**2
    return epsilon/2 * integrate(inside_integral, a, T) + (1-epsilon)/2*np.linalg.norm(X_t(T, v))**2
    #*||X(T)||**2
```

####Résolution de $inf_v J(v)$

```python
def Gradient_descent_fixed(func, p, x0, N, gradient=gradient,
                           epsilon=0.001, Iterations = 4004, plot = False):
    # initialisation
    x_old = np.random.normal(size = N)
    x_new = x0
    i = 0
    routeF = np.array([func(x0)])
    routeX = np.array([x0])
    
    # la boucle principal
    while np.linalg.norm(x_old-x_new) > epsilon and i < Iterations:
        x_old = np.copy(x_new)           
        d = -1*gradient(func, x_old)
        x_new = x_old + p*d
        
        routeF = np.append(routeF, func(x_new))
        routeX = np.concatenate((routeX, [x_new]))
        i+=1
    
    # plot pour voir la route du gradient
    if plot:
        i+=1
        # on calcule les erreurs sur chaque pas
        errors = np.zeros(i)
        for k in range(i):
            errors[k] = np.linalg.norm(routeX[k] - x_new)      
        
        xx = np.arange(0, i, 1)
        
        fig,axs = plt.subplots(2, 1, figsize = (12, 6), dpi =128, sharex = True)
        axs[0].set_title("Valeurs de la fonction"
                         " évaluée à chaque pas de gradient à pas fixe")
        axs[0].plot(xx, routeF, "--", color = "coral", linewidth = 2)
        axs[1].set_title("Erreurs - Log scale")
        axs[1].set_yscale("log")
        axs[1].plot(xx, errors, "--", color = "teal", linewidth = 2)
        plt.xlabel("Iteration");
    
    return x_new, routeF, routeX
```
