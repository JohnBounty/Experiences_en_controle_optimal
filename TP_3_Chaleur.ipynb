*Contrôle optimal de l'équation de la chaleur*

- - -

On considère une barre unidimensionnelle de longueur $L > 0$. 

La température de l’extrémité gauche de la barre ( en $x = 0$) est maintenue 
à $0$  $^{o}K$ et on cherche à contrôler la température de la barre à l’extrémité droite (en $x = L$) à l’aide d’un flux de chaleur $t 􏰂\rightarrow u(t)$. On appelle $y_u(t, x)$ la température de cette barre au temps $t \in [0, T ]$ et en $x \in [0, L]$.

On suppose donc que $y_u$ résout l’équation de la chaleur:

$$
(1)\quad\quad    \begin{cases}{\partial_{t} y_{u}(t, x)-\partial_{x x} y_{u}(t, x)=0} & {(t, x) \in] 0, T[\times] 0, L[} \\ {y_{u}(0, x)=0} & {x \in[0, L]} \\ {y_{u}(t, 0)=0} & {t \in[0, T] \quad \text { (conditon de Dirichlet homogène) }} \\ {\partial_{x} y_{u}(t, L)=u(t)} & {t \in[0, T] \quad \text { (conditon de Neumann non-homogène) }}\end{cases}
$$

et on s’intéresse au problème de contrôle optimal:

$$
\mathcal{(P)} \quad\quad \inf _{u \in L^{2}(] 0, T[)} J(u) \quad \text { où } J(u)=\frac{1}{2} \int_{0}^{T}\left(y_{u}(t, L)-z_{d}(t)\right)^{2} d t+\frac{\varepsilon}{2} \int_{0}^{T} u(t)^{2} d t \\
$$


où $\varepsilon > 0$ est un paramètre fixé et $z_{d} \in L^{2}(|0, T|)$ est donnée.

## I. On souhaite montrer que J est différentiable.


L'application $u \rightarrow J(u)$ se compose de deux termes, qui sont la distance $L_2$ entre la solution en $L$ et la cible, et le coût du contrôle, représenté par la norme $L_2$. La somme de deux fonctionnelles différentiables reste différentiable.

De plus, l'application $u \rightarrow y_u$ est linéaire et continue; on suppose et accepte qu'elle est également continue en point d'espace $L$... alors, $J(u)$ est différentiable.

Calculons le $\nabla J(u)$!

On commence par noter le différentiel de $J(u)$ en $u$ dans la direction $(h)$:

$$
    DJ_u(h) = \lim_{\varepsilon \rightarrow 0} \frac{J(u+\varepsilon h) - J(u)}{\varepsilon} = \int_0^T \hat{y}(y_u(t, L) - z_d)dt + \varepsilon\int_0^T u(t)h(t)dt, \quad     
    \text{où } \hat{y} \text{ résout: } \\
    (\chi)\quad\quad \begin{cases}\partial_t \hat{y}(t, x) - \Delta \hat{y}(t, x) = 0 \\ 
    \hat{y}(0, x) = 0 \\
    \hat{y}(t, 0) = 0 \\
    \partial_x\hat{y}(t, L) = h \end{cases} \\
$$

On introduit le problème adjoint:

$$
    (\xi)\quad\quad \begin{cases} -\partial_t p(t, x) - \Delta p(t, x) = 0 \\ 
    p(0, x) = \quad ? \\
    p(t, 0) = \quad ?  \\
    \partial_x p(t, L) =\quad ? \end{cases} \\
$$

On va appeler l'espace de travail en espace $\Omega = [0, L]$ pour simplifier les notations dans la suite. Il faut remarquer que dans notre cas (dimension = 1), $\partial\Omega = {0, L}$. On va également noter $\hat{y}(t, x) = \hat{y}$ et $\int_a^b f(t, x) dxdt$ comme $\int_a^b f(t, x)$.

On multiplie $(\chi)$ par $p$ et on intègre en espace et en temps:

$$
    \int_{\Omega}\int_0^T \partial_t \hat{y} p - \int_{\Omega}\int_0^T \Delta \hat{y} p = 0 \\
$$

En appliquant I.P.P. sur $\Delta \hat{y}$, on obtient:

$$
    \int_{\Omega}\int_0^T \partial_t \hat{y} p + \int_{\Omega}\int_0^T \nabla \hat{y} \nabla p 
    -  \int_{\partial\Omega}\int_0^T \partial_n \hat{y} p = 0 \\
$$

Maintenant, on multiplie $(\xi)$ par $\hat{y}$ et on intègre en espace et en temps:

$$
    -\int_{\Omega}\int_0^T \partial_t p \hat{y} - \int_{\Omega}\int_0^T \Delta p \hat{y} = 0 \\
$$

En appliquant I.P.P. sur les deux termes, on obtient:

$$
    \int_{\Omega}\int_0^T \partial_t \hat{y} p - \int_{\Omega}\left[ p(t,x) \hat{y}(t,x) \right]_{t=0}^{t=T} + \\
    + \int_{\Omega}\int_0^T \nabla \hat{y} \nabla p 
    -  \int_{\partial\Omega}\int_0^T \partial_n p \hat{y} = 0, \quad \text{ où } \\
    \int_{\partial\Omega}\int_0^T \partial_n p \hat{y} = \int_0^T \partial_x p(t, L) \hat{y}(t, L) - \underbrace{\int_0^T \partial_x p(t, 0) \hat{y}(t, 0)}_{ = 0, \text{ car } \hat{y}(t, 0) = 0}
$$

En considérant les deux, on peut écrire l'identité de dualité:

$$
    \underline{ \int_{\partial\Omega}\int_0^T \partial_n \hat{y} p = \int_{\Omega} p(T,x)\hat{y}(T, x) + \int_0^T \partial_x p(t, L) \hat{y}(t, L) } \quad \text{ où }\\
    \int_{\partial\Omega}\int_0^T \partial_n \hat{y} p = \int_0^T \underbrace{\partial_x \hat{y}(t, L)}_{=h} p(t, L) - \int_0^T \partial_x \hat{y}(t, 0) p(t, 0)
$$

Rappelons, qu'on veut mettre $\int_0^T \hat{y}(y(t, L) -z_d)$ sous la forme $\int_0^T h \times truc$. Alors, on pose:

$$
    p(T, x) = 0 \\
    p(t, 0) = 0 \\
    \partial_x p(t, L) = y(t, L) - z_d \\
$$

Ce, qui donne:

$$
    \int_0^T \hat{y}(y(t, L) - z_d) = \int_0^T h p(t, L) \\
    \text{et, donc,} \quad truc = p(t, L)
$$

Enfin, on peut écrire le gradient sous la forme suivante:

$$
    \boxed{\quad \nabla J(u) = p(t, L) + \varepsilon u(t)\quad } \\
$$

où $p$ est l'état adjoint de $(\chi)$, solution de:

$$
    (\xi)\quad\quad \begin{cases} -\partial_t p(t, x) - \Delta p(t, x) = 0 \\ 
    p(T, x) = 0 \\
    p(t, 0) = 0  \\
    \partial_x p(t, L) = y(t, L) - z_d \end{cases} \\
$$

## II. Pour résoudre numériquement le problème étudié on propose une méthode de différences finies avec un schéma implicite.

### Description de la méthode

On introduit une discrétisation pour $y_u(t,x)$ à la fois dans le temps et dans l'espace. A chaque instant du temps $k$, en chaque point de la barre $i$ on calcule $y_i^k$ de la façon suivante :

$$
    y_i^k \simeq y(k\Delta t, i \Delta x), 
$$

où $\Delta t$ est un pas de discrétisation uniforme en temps, $\Delta x$ est un pas de discrétisation uniforme en espace. On définit, alors, $t_k = k\Delta t$ et $x_i = i \Delta x$. On note $N$ - le nombre des pas d'espace, $K$ - nomre des pas du temps.

Ainsi, les dérivés de $y$, $\partial_{t} y_{u}(t, x)$ et $\partial_{x x} y_{u}(t, x)$, présents dans le (1) peuvent être réécrits et calculés de la manière suivante:

$$
 \frac{\partial^2 y}{\partial x^2} (t_k, x_i) \simeq \frac{y_{i+1}^k + y_{i-1}^k - 2y_i^k}{(\Delta x)^2}, \quad\text{et} \\
\frac{\partial y}{\partial t} (t_k, x_i) \simeq \frac{y_i^k - y_i^{k-1}}{\Delta t} \quad \quad
$$



Ce qui nous permet de mettre l'équation principale sous la forme discrétisée :

$$
    \frac{y_i^k - y_i^{k-1}}{\Delta t} - \frac{y_{i+1}^k + y_{i-1}^k - 2y_i^k}{(\Delta x)^2} = 0 \\
$$

Les conditions au bord sont les suivantes:

$$
    y_i^0 = 0, \quad\forall i = 1..N \\
    y_0^k = 0, \quad\forall k = 1..K \quad \text{(condition de Dirichlet homogène)} \\
    y^k_{L} = y^k_{N} + u(t^k) \Delta x \quad\forall k = 1..K \quad \text{(condition de Neumann non-homogène)} \\
$$



Le système se réécrit sous la forme matricielle avec:

$$
    Y^k = \begin{pmatrix} y^k_1 \\ \vdots \\ y^k_{N+1}\end{pmatrix}, \quad 
    b^k = \begin{pmatrix} 0 \\ \vdots \\ u(t^k)\end{pmatrix} \\
    A = \begin{pmatrix} 2 & -1 & 0 & \dots & 0\\ 
    -1 & 2 & -1 & \dots & 0\\  
    \vdots & \ddots& \ddots & \ddots & \vdots \\ 
    0 & 0 & \ddots & 2 & -1\\
    0 & 0 & \dots & -1 & 1\\    \end{pmatrix} \\
$$

Ce qui permet d'écrire l'équation principale sous la forme :

$$
    Y^k - Y^{k-1} + \frac{\Delta t}{(\Delta x)^2}AY^k = \frac{\Delta t}{\Delta x} b^k \implies \boxed{\left( I + \frac{\Delta t}{(\Delta x)^2}A \right)Y^k = \frac{\Delta t}{\Delta x}b^k + Y^{k-1}}
$$

### La résolution numérique


```python
# imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# les parametres de descritisation 
Nx = 100
Nt = 80
L = 1
T = 10

dt = T/Nt
dx = L/Nx

def u_func(t):
    return t**2 + 3
```

```python
x = np.linspace(0, L, Nx+1)
t = np.linspace(0, T, Nt+1)
```

```python
def make_A(N, k1=-1, k2=2, k3=-1):
    a = np.ones(N-1) * k1
    b = np.ones(N) * k2
    c = np.ones(N-1) * k3
    return np.diag(a, -1) + np.diag(b, 0) + np.diag(c, 1)

def make_b(Neumann, N):
    b = np.zeros(N)
    # la condition de Neumann
    b[-1] = Neumann
    return b
```

```python
# On inverse la partie gauche de l'équation principale avant de commencer la boucle
# Elle ne change pas, alors on doit gagner en termes de l'efficacité
A = make_A(Nx+1)
left_side = np.eye(Nx+1) + (dt/(dx**2))*A
A1 = np.linalg.inv(left_side)
```

```python
def one_step(yp, LS, bk, dt, dx):
    # LS - partie gauche
    # yp - y du pas de temps précédent
    rhs = yp + (dt/dx)*bk
    yn = LS@rhs
    return yn
```

```python
res = np.empty((Nx+1, Nt+1))
y0 = np.zeros(Nx+1)
yp = y0.copy()
res[:, 0] = y0

for time_step in range(1, Nt+1):
    uk = u_func(time_step)
    bk = make_b(uk, Nx+1)
    yn = one_step(yp, A1, bk, dt, dx)
    res[:,time_step] = yn
    yp = yn.copy()
```

```python
# Un peu de vérification de la pertinence physique :
# 0 degrés Kelvin est un zéro absolu, 
# donc on ne peut pas atteindre une température négative dans cette expérience. 
# Vérifions si elle tient !

print("Any negatives ? ", res[res<0])
```

```python
fig, ax = plt.subplots(1, 1, figsize = (16, 7))
plt.title("L'Évolution de la température dans la barre ", fontsize = 20)
ax = sns.heatmap(res.T, cmap ='hot',
            xticklabels = 10, yticklabels = 10);
ax.set_ylabel("Le temps", fontsize = 14);
ax.set_xlabel("L'espace", fontsize = 14);
        
```

```python
n_plots = 10
chosen_ts = np.arange(0, Nt+1, (Nt+1)//n_plots)
fig, axs = plt.subplots(n_plots+1, 1, figsize = (16, 0.5*n_plots))
fig.suptitle("L'Évolution de la température dans la barre - II", fontsize = 20)
for i in range(n_plots+1):
    res_plot = np.transpose(res[:, chosen_ts[i], np.newaxis])
    axs[i].imshow(res_plot, cmap ='hot', interpolation='nearest');
    axs[i].set_xticklabels([''])
    axs[i].set_yticklabels(['t = {:d}'.format(chosen_ts[i])])
```

***Commentaire:*** Il n'est pas si évident de visualiser ce genre de données, mais, une "heatmap" est, certainement, un bon choix pour visualiser les résultats de la modélisation de la chaleur.

Sur cet exemple simple avec $u(t) = t^2 + 3$ étant monotone de $0$ à $T$, on peut observer comment la température augmente avec le temps. On voit clairement que les valeurs sont plus élevées sur le côté droit de la barre, là où se trouve la source de chaleur. Ainsi, on confirme que le résultat obtenu est bien attendu et logique.

## III. Le temps renversé

On pose $q(t, x) = p(T-t, x)$. Alors, le système $(\xi)$ devient:

$$
    \begin{cases} \partial_t q(t, x) - \Delta q(t, x) = 0 \\ 
    q(0, x) = 0 \\
    q(t, 0) = 0  \\
    \partial_x q(t, L) = y(t, L) - z_d \end{cases} \\
$$

Ce, qui vas nous aider résoudre le problème $\mathcal{(P)}$.

## IV. Résolution numérique du problème $\mathcal{(P)}$

***Commentaire:*** Pour résoudre numériquement ce problème, on considérera la même approche que dans le TP 2. $J(u)$ sera optimisé en utilisant la descente de gradient à pas constant $\rho > 0$ ; deux équations différentielles sont résolues à chaque itération : une pour l'état et une pour son adjoint. Cependant, cette fois-ci, pour résoudre ces équations différentielles, on utilisera une méthode décrite en II.

### Encapsulation de la résolution de l'équation différentielle

```python
def solve_eq(Nx, Nt, dx, dt, A1, NeumannC, plot=False):
    res = np.empty((Nx+1, Nt+1))
    y0 = np.zeros(Nx+1)
    yp = y0.copy()
    res[:, 0] = y0

    for time_step in range(1, Nt+1):
        nc = NeumannC[time_step]
        bk = make_b(nc, Nx+1)
        yn = one_step(yp, A1, bk, dt, dx)
        res[:,time_step] = yn
        yp = yn.copy()

    if plot:
        fig, ax = plt.subplots(1, 1, figsize = (16, 7))
        plt.title("L'Évolution de la température dans la barre ", fontsize = 20)
        ax = sns.heatmap(res.T, cmap ='hot',
                    xticklabels = 10, yticklabels = 10);
        ax.set_ylabel("Le temps", fontsize = 14);
        ax.set_xlabel("L'espace", fontsize = 14);

    return res
```

```python
# test
utest = u_func(np.arange(0, Nt+1))
res = solve_eq(Nx, Nt, dx, dt, A1, utest, plot=True);
```

### Résolution du $\mathcal{(P)}$

```python
def J(u, y_L, zd, Nt, eps):
    return (1 / (2 * Nt)) * np.sum((y_L - zd)**2) + (eps / (2 * Nt)) * np.sum(u**2)

def dJ(u, p_L, T, eps):
    return p_L + eps * u
```

```python
def inf_J(ro, u0, Nx, Nt, T, L, J, dJ, A, zd, eps,
          epsilon=1e-5, print_epoch=25, 
          Iterations = 4004, plot = False,
          cooling_epoch=1000000, heating_epoch = 1000000):

    # initialisation
    # discrétisation
    dt = T/Nt
    dx = L/Nx
    # matrice A1 pour la résolution
    A = make_A(Nx+1)
    left_side = np.eye(Nx+1) + (dt/(dx**2))*A
    A1 = np.linalg.inv(left_side)
    # contrôle
    u_old = np.random.normal(Nt+1)
    u_new = u0
    # état
    ys = solve_eq(Nx, Nt, dx, dt, A1, u0)
    # adjoint
    neumann = ys[-1, :]
    neumann = neumann[::-1] - zd
    ps = solve_eq(Nx, Nt, dx, dt, A1, neumann)
    i = 0

    routeF = np.array([J(u0, ys[-1, :], zd, Nt, eps)])
    routeX = np.array([u0])
    
    try:
        # la boucle principal
        while np.linalg.norm(u_old-u_new) > epsilon and i < Iterations:
            i+=1
            u_old = np.copy(u_new)
            # resolution du (1)
            ys = solve_eq(Nx, Nt, dx, dt, A1, u_old)
            neumann = ys[-1, :]
            neumann = neumann[::-1] - zd
            # resolution du (2)
            ps = solve_eq(Nx, Nt, dx, dt, A1, neumann)
            
            # pas du gradient
            d = -1*dJ(u_old, ps[-1, :], T, eps)
            u_new = u_old + ro*d
            
            routeF = np.append(routeF, J(u_new, ys[-1, :], zd, Nt, eps))
            routeX = np.concatenate((routeX, [u_new])) 

            # les expériences avec
            # changement de la vitesse de descente du gradient
            if i%cooling_epoch == 0:
                ro /= 10

            if i%heating_epoch == 0:
                ro *= 10     
            
            if i%print_epoch == 0:
                print("[Minimization epoch {:d}] >>> target function value is {:.4f}.".format(i, routeF[i]))
            

    except KeyboardInterrupt:
        print("The algorithm was stopped at epoch {:d}.".format(i))
    
    # min_iter = np.argmin(routeF[1:])
    # u_opt = routeX[min_iter]
    u_opt = u_new.copy()
    # plot pour voir la route du gradient
    if plot:
        i = len(routeF)
        # on calcule les erreurs sur chaque pas
        errors = np.zeros(i)
        for k in range(i-1):
            errors[k] = np.linalg.norm(routeX[k] - u_opt)      
        
        
        xx = np.arange(0, i, 1)
        # print("F: ", routeF.shape)
        # print("xx: ", xx.shape)
        # print("Errs: ", errors.shape)
         
        fig,axs = plt.subplots(2, 1, figsize = (12, 6), dpi =128, sharex = True)
        axs[0].set_title("Valeurs de la fonctionnelle"
                         " évaluée à chaque pas de gradient à pas fixe")
        axs[0].plot(xx, routeF, "--", color = "coral", linewidth = 2)
        #axs[0].plot(min_iter, routeF[min_iter], "*", markersize = 5, color = "orange")
        axs[1].set_title("Erreurs - Log scale")
        axs[1].set_yscale("log")
        axs[1].plot(xx, errors, "--", color = "teal", linewidth = 2)
        plt.xlabel("Iteration");
    
    return u_opt, routeF, routeX, ys, ps
```

### Expérience A

```python
# les parametres pour verifier
Nx = 100
Nt = 100
L = 1
T = 1 # ou 1 ou 10
zd = 1 #partout
eps = 1e-3
dt = T/Nt # 0.001 ou 0.0001
dx = L/Nx
ro = 5
```

```python
u_opt, routeF, routeX, ys, ps = inf_J(ro, u0, Nx, Nt, T, L, J, dJ, A, zd, eps,
                                      epsilon = 1e-10, heating_epoch = 2000, 
                                      print_epoch = 100,
                                      Iterations = 30000, plot=True)
```

## Conclusion

*On peut observer que la fonctionelle $J(u)$ est optimisée, mais peu de vérifications simples montrent que la solution obtenue est en fait une constante proche de la cible $z_d$ : dans l'expérience A la solution obtenue dans $L$ est 11 fois plus petite que $z_d$. On en conclut donc qu'il y a probablement un bug quelque part dans l'implémentation proposée de la résolution du problème $\mathcal{(P)}$.*

*Néanmoins, on a étudié le problème de la modélisation de la chaleur dans la barre unidimensionnelle dans le temps. La méthode des différences finies avec schéma implicite a été utilisée pour résoudre l'équation différentielle. À un moment donné, il faut résoudre le problème du temps rétrograde, et c'est pourquoi le schéma implicite est très pratique.  On a aussi montré comment les méthodes de l'adjoint sont utilisées pour calculer le gradient de $J(v)$ et résoudre le problème de minimisation correspondant.*
